# Seaborn
* Matplotlib의 기능과 스타일을 확장한 파이썬 시각화 도구의 고급 버전
* 비교적 단순한 인터페이스를 제공

출처 : '파이썬 머신러닝 판다스 데이터 분석'

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.info()
```

## Line plot

```python
# 스타일 테마 설정(5가지 : darkgrid, whitegrid, dark, white, ticks)
sns.set_style('darkgrid')
sns.lineplot(x='age', y='fare', data=titanic)
plt.show()
```

## Bar chart

```python
sns.barplot(x='sex', y='survived', data=titanic)
plt.show()
```

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
plt.show()
```

```python
sns.barplot(x='sex', y='survived', hue='class', dodge=False, data=titanic)
plt.show()
```

## Histogram

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# distplot
sns.distplot(titanic['fare'], ax=ax1)

# kdeplot
sns.kdeplot(titanic['fare'], ax=ax2)

# histplot
sns.distplot(titanic['fare'], kde=False, ax=ax3)

plt.show()
```

## scatter plot

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

# 회귀선이 있는 산점도
sns.regplot(x='age', y='fare', data=titanic, fit_reg=True, ax=ax1)

# 회귀선이 없는 산점도
sns.regplot(x='age', y='fare', data=titanic, fit_reg=False, ax=ax2)

plt.show()
```

## Box plot

```python
sns.boxplot(x='alive', y='age', data=titanic)
plt.show()
```

```python
sns.boxplot(x='alive', y='age', hue='sex', data=titanic)
plt.show()
```

```python
# 바이올린 플랏
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.violinplot(x='alive', y='age', data=titanic, ax=ax1)
sns.violinplot(x='alive', y='age', hue='sex', data=titanic, ax=ax2)
plt.show()
```

## Heatmap

```python
df = titanic.corr()
df
```

```python
# 그림 사이즈 지정
fig, ax = plt.subplots( figsize=(7,7) )

# 삼각형 마스크를 만든다(위 쪽 삼각형에 True, 아래 삼각형에 False)
mask = np.zeros_like(df, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# 히트맵을 그린다
sns.heatmap(df, 
            cmap = 'RdYlBu_r', 
            annot = True,   # 실제 값을 표시한다
            mask=mask,      # 표시하지 않을 마스크 부분을 지정한다
            linewidths=.5,  # 경계면 실선으로 구분하기
            cbar_kws={"shrink": .5},# 컬러바 크기 절반으로 줄이기
            vmin = -1,vmax = 1   # 컬러바 범위 -1 ~ 1
           )  
plt.show()
```

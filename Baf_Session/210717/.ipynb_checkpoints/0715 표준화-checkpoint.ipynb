## 데이터셋 분할
### data19 : 2015~2019
### data20 : 2020~2021

```python
data19 = data.loc[data['year'] < 2020,:]
data19
```

```python
data20 = data.loc[data['year'] >= 2020,:]
data20
```

## 표준화 함수

```python
def standard(data) :
    result = (data - np.mean(data, axis=0)) / np.std(data, axis=0)
    return result
```

## 특정 칼럼 표준화

```python
def column_standard(col) :
    col_new = []
    for y in range(2015, 2022) : 
        data_s = data.loc[data['year'] == y,:]
        x = standard(data_s[col]).to_list()
        col_new = col_new + x
    return col_new
```

## social_support	health	freedom	trust	generosity	dystopia 표준화

```python
gdp_std = []
for y in range(2015, 2022) :
    if y < 2020 :
        data_s = data19.loc[data['year'] == y,:]
        x = standard(data_s['gdp']).to_list()
        gdp_std = gdp_std + x
    else :
        data_s = data20.loc[data['year'] == y,:]
        x = standard(data_s['Logged GDP per capita']).to_list()
        gdp_std = gdp_std + x
data['gdp_std'] = gdp_std
data
```

```python
data['social_support_std'] = column_standard('social_support')
```

```python
data['health_std'] = column_standard('health')
```

```python
data['freedom_std'] = column_standard('freedom')
```

```python
data['trust_std'] = column_standard('trust')
```

```python
data['generosity_std'] = column_standard('generosity')
```

```python
data['dystopia_std'] = column_standard('dystopia')
```

```python
data['score_std'] = column_standard('score')
```

```python
import matplotlib.pyplot as plt
```

```python
# Scatter plot with a different color by groups
def scat(col1,col2) :
    groups = data.groupby('year')

    fig, ax = plt.subplots()
    for year, group in groups:
        ax.plot(group[col1], 
               group[col2], 
               marker='o', 
               linestyle='',
               label=year)

    ax.legend(fontsize=12, loc='upper left') # legend position
    plt.title(col1)

    plt.show()
```

```python
scat('social_support','social_support_std')
```

```python
scat('health','health_std')
```

```python
scat('freedom','freedom_std')
```

```python
scat('trust','trust_std')
```

```python
scat('generosity','generosity_std')
```

```python
scat('dystopia','dystopia_std')
```

```python
scat('gdp','gdp_std')
```

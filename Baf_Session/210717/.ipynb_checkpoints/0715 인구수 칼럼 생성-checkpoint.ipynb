```python
from pypapago import Translator
# 번역할 파일명
fileName = './test.txt'
# 사용할 전역변수
templist = []
afterTranslate = []
# 객체 할당
translator = Translator()
# 파일 읽기모드, 줄을 읽고 \n를 제거하여 번역하기
with open(fileName, 'r') as f: 
    lines = f.readlines()
    for line in lines:
        # '\n'엔터키''가 있을 때 교체하여 추가, 그 외 그냥 추가
        if '\n' in line:
            line = line.replace('\n', '')
            templist.append(line)
        else:
            templist.append(line)
        # 문자열이 비어있는 리스트는 삭제
        try:
            templist.remove('')
        except:
            pass
# 리스트 내 영어 자료들 번역하기
for i in range(len(templist)):
    result = translator.translate(templist[i], source='en', target='ko', verbose=False)
    afterTranslate.append(result)
```

```python
translator = Translator()
afterTrans = []

for d in data['국가별'] :
    trans = translator.translate(d, src='kr', tgt='en').lower()
    afterTrans.append(trans)

afterTrans
```

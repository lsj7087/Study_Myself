#### 1. 데이터 불러오기 및 확인

#### 그 외 : apply()

```python
# 우선 lambda함수의 구조를 간단히 파악해봅시다
# lambda함수는 사용자지정의 함수를 문법에 맞춰 작성하는 것 보다 더 간단하게 한줄로 해결가능

def func1(x,y):
    return x+y

func2 = lambda x,y : x+y

print('func1 : ', func1(7,10))
print('func2 : ', func2(7,10))

# 두 함수 모두 결과는 같지만 lambda함수의 구조가 훨씬 간단하다
```

```python
## apply()는 lambda함수와 연결해 사용한다.
## apply()를 통해 새로운 변수를 생성해보자

stock['Diff'] = stock.apply(lambda x : x['Close'] - x['Open'], axis = 1)

stock
```

```python
# if 조건 추가

stock['EUR Diff'] = stock.apply(lambda x: x['High'] - x['Low']
                               if (x['Currency'] == 'EUR')
                               else -1, axis = 1)

stock
```

```python
stock['EUR Diff'].describe()
```
